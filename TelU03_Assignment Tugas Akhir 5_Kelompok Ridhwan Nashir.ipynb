### Memproses Jenjang Sekolah 

```python
datapendidikan['Jenjang Sekolah'].unique()
```

Dari data unique value diatas, maka data jenjang sekolah akan dibagi menjadi PAUD/TK, SD, SMP, SMA/SMK, SLB, dan PT (perguruan tinggi). Penyesuaian akan mengikuti kelompok-kelompok tadi.

```python
datapendidikan['Jenjang Sekolah'].replace(['PAUD','TK','PAUD/TK/KOBER','TK/RA','KOBER/PAUD','TK ','KOBER','KOBER ','PAUD TPA','TK dan PAUD','TKQ','PAUD ','Kober ','TPA','TK / RA / DTA'],'PAUD/TK',inplace=True)
datapendidikan['Jenjang Sekolah'].replace(['SD  ','SD/MI','SD ','SDN','MI','Madrasah'],'SD',inplace=True)
datapendidikan['Jenjang Sekolah'].replace(['SMP/MTS','SMPN','MTs'],'SMP',inplace=True)
datapendidikan['Jenjang Sekolah'].replace(['SMK','SMA/SMK/MA','SMA','SMAN','SMA ','SMAK'],'SMA/SMK',inplace=True)
datapendidikan['Jenjang Sekolah'].replace(['Universitas','UNIVERSITAS','Perguruan Tinggi','Univ'],'PT',inplace=True)
```

### Memproses Data Masyarakat Miskin 

Berdasarkan eksplorasi data masyarakat miskin, terdapat 2 fitur utama yang dibutuhkan untuk membangun model yaitu Kelurahan dan Jumlah Warga Miskin. <br>
Namun, dalam dataset tersebut jumlah warga miskin terpecah menjadi beberapa kolom yang membuat banyak data null, sehingga perlu disatukan terlebih dahulu.

```python
datamiskin.replace('-',None,inplace=True)
datamiskin.replace(' -',None,inplace=True)
datamiskin.replace(' - ',None,inplace=True)
datamiskin
```

```python
datamiskin.fillna(value=int(0),inplace=True)
dataset = pd.DataFrame(datamiskin['Kelurahan'])
dataset['Jumlah Warga Miskin'] = datamiskin['Jumlah Anggota Keluarga'].astype(int) + datamiskin['Jumlah Anggota Keluarga Miskin'].astype(int) + datamiskin['Jumlah Masyarakat Miskin Anggota Keluarga'].astype(int) + datamiskin['Jumlah '].astype(int) + datamiskin['Anggota Keluarga Masyarakat Miskin'].astype(int) + datamiskin['Jumlah Anggota Keluarga Masyarakat Miskin'].astype(int) + datamiskin[' Jumlah Anggota Keluarga Miskin '].astype(int)
dataset
```

```python
dataset.groupby('Kelurahan').sum()
```

Dari tabel jumlah warga miskin, terdapat beberapa kelurahan yang berjumlah 0 karena dataset yang kurang baik. <br>
Kelurahan tersebut (Cicadas, Cigadung, Neglasari, Taman Sari) akan di drop. Kemudian menambahkan jenjang sekolah

```python
# Drop kelurahan pada dataset 
dataset = dataset.groupby('Kelurahan').sum()
dataset = dataset.loc[(dataset!=0).any(axis=1)]
dataset = dataset.reset_index()

# Drop kelurahan pada dataset 
datapendidikan = datapendidikan[(datapendidikan.Kelurahan != 'Cicadas')&(datapendidikan.Kelurahan != 'Cigadung')&(datapendidikan.Kelurahan != 'Neglasari')&(datapendidikan.Kelurahan != 'Taman Sari')]

# Add jenjang sekolah
jenjangsekolah = datapendidikan['Jenjang Sekolah'].unique()
for jenjang in jenjangsekolah:
     dataset[jenjang] = 0
dataset
```

```python
# Menghitung jumlah sarana pendidikan dari masing-masing kelurahan
jumlahsarana = datapendidikan.groupby('Kelurahan').size()
jumlahsarana
```

```python
# Menghitung data sarana pendidikan berdasarkan kelurahan
kelurahan = dataset['Kelurahan'].values
for i in range(len(kelurahan)):
    for index, row in datapendidikan.iterrows():
        if row['Kelurahan'] == kelurahan[i]:
            if row['Jenjang Sekolah'] == 'PAUD/TK':
                dataset['PAUD/TK'].iloc[i] += 1
            if row['Jenjang Sekolah'] == 'SD':
                dataset['SD'].iloc[i] += 1
            if row['Jenjang Sekolah'] == 'SMP':
                dataset['SMP'].iloc[i] += 1
            if row['Jenjang Sekolah'] == 'SMA/SMK':
                dataset['SMA/SMK'].iloc[i] += 1
            if row['Jenjang Sekolah'] == 'SLB':
                dataset['SLB'].iloc[i] += 1
            if row['Jenjang Sekolah'] == 'PT':
                dataset['PT'].iloc[i] += 1
```

```python
# menambah fitur jumlah sarana
dataset['Jumlah Sarana'] = jumlahsarana.values
dataset = dataset.sort_values(by='Jumlah Warga Miskin').reset_index(drop=True)
dataset
```
